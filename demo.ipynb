```python
from detectron2.data import build_detection_test_loader
from detectron2.data import MetadataCatalog, DatasetCatalog
import cv2, random
from detectron2.utils.visualizer import Visualizer
import matplotlib.pyplot as plt
from detectron2.data.datasets import register_coco_instances
%matplotlib inline


register_coco_instances("ade20k2021_train", {}, "/project/n/p10922001/DoorSeg/datasets/ADE20K_2021_17_01/ade20k_instance_train.json", "/project/n/p10922001/DoorSeg/")

dataset = DatasetCatalog.get("ade20k2021_train")
for d in random.sample(dataset, 3):
    img = cv2.imread(d["file_name"])
    # raise
    visualizer = Visualizer(img[:, :, ::-1], metadata=MetadataCatalog.get("ade20k2021_train"))
    out = visualizer.draw_dataset_dict(d)
    plt.title(d["file_name"].split('/')[-1])
    plt.imshow(cv2.cvtColor(out.get_image()[:, :, ::-1], cv2.COLOR_BGR2RGB))
    plt.show()
```
